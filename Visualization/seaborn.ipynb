### Distribution Plot

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'])
```

```python
# The line that we got is called kernal density estimation (KDE)
sns.distplot(tips['total_bill'],kde=False)
```

```python
sns.distplot(tips['total_bill'],kde=False, bins=30)
```

```python
sns.distplot(tips['total_bill'],kde=False, bins=100)
# Note x axis represent total amount of bill. and y axis represent the total no of bills
```

### Joint Plot

```python
# it allows to combine two distplots for bivariate data
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
# hex for hexagon 
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
# reg is for regression 
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

### Pairplot

```python
# Does for every column every combination
sns.pairplot(tips)
```

```python
# categories can be passed as hue
sns.pairplot(tips, hue='sex')
```

```python
sns.pairplot(tips, hue='sex', palette='coolwarm')
```

### Rug Plot

```python
# Good to know the frequency density. | for a single point in distribution
sns.rugplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'], kde=False)
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.rugplot(tips['total_bill'])
```

```python
# Note that rugplot is used to create the KDE line that we saw in distplot!
```

```python
sns.kdeplot(tips['total_bill'])
```
