```python
import seaborn as sns
%matplotlib inline
iris = sns.load_dataset('iris')
iris.headd()
```

```python
# It has categories
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
# pairgrid just creates the grids with all possible combination (bi variate ) but unlike pairplot doesn't plot anything
sns.PairGrid(iris)
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

### Facet Grid

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
g = sns.FacetGrid(data=tips, col='time',row='smoker')
g.map(sns.distplot, 'total_bill')
```

```python
g = sns.FacetGrid(data=tips, col='time',row='smoker')
g.map(plt.scatter, 'total_bill', 'tip')
```
