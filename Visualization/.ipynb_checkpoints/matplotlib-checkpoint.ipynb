```python
import matplotlib.pyplot as plt
```

```python
# allows plots to show up inside notebook
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
x
```

```python
y = x**2
y
```

```python
# 2 ways to create matplotlib plot
```

```python
# functional 
plt.plot(x,y)
```

```python
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
plt.plot(x,y)
```

```python
# multi plot
# subplot args (#rows. #cols, index)
# r and b is for color
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object Oriented API Method (USe this)
```

```python
# fig is like a blank canvas
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
axes1.plot(x,y)
axes1.set_title('Larger Plot')
axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

```python
"""REVIEW: Whenever we create an empty canvas object by canvasObj = plt.figure(). it's just a blank canvas, on that we can add axes, on which we can plot something, we can have multiple axes"""
```

### Subplots

```python
fig,axes = plt.subplots()
axes.plot(x,y)
```

```python
# tuple unpack fig: canvas obj, axes = list of axes(subplots)
fig,axes = plt.subplots(nrows=1, ncols=2)
print(type(axes))
print(type(fig))
# we can see that axis is list of axes
#so we can do 
for sub_axes in axes:
    sub_axes.plot(x,y)
# This plt.subplot is doing what we did before (i.e, fig.add_axes) automatically by specifying nrows and ncols
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y)
axes[0].set_title('y = x^2')
axes[1].plot(y,x)
axes[1].set_title('x = y^2')
```

## Figure Size and DPI

```python
# note fig size is in inches, dpi indicate dotsperinch (greater => clear)
fig = plt.figure(figsize=(8,2), dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2,figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)
# tight_layout tries best to fit the subplots 
plt.tight_layout()
```

### Legends and other titles and multiple graph

```python
canvsObj = plt.figure()
subPlt = canvsObj.add_axes([0,0,1,1])
subPlt.plot(x,y, label='Y = X^2')
subPlt.plot(x,y+10, label='Y = X^2 + 10')
subPlt.set_xlabel('X axis')
subPlt.set_ylabel('Y axis')

#subPlt.legend(loc='center')
subPlt.legend(loc=0)
# we can also put the location of legend, set it as 0 for best
```

# Customising Plots

```python
#color
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='cyan')
```

```python
# line width and style
# linewidth = fraction of default (eg 2 => double of default) we can use lw= or linewidth=
# alpha => transparency
# linestyle = style we can use ls= or linestyle=
# Markers (to mark the points) marker='+'/'o'/'*'
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='red', linewidth=2, alpha=0.5, linestyle='-.', marker='o', markersize=10, markerfacecolor='Green')
```

```python
# Controling axis scale

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='red', linewidth=2, alpha=0.5, linestyle='-.', marker='o')

ax.set_xlim([0,4])
ax.set_ylim([0,4])
```
