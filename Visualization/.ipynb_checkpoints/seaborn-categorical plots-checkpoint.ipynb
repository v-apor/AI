```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

#### BarPlot


```python
# aggregate by some function (default is mean), it's like viz of groupby
sns.barplot(x='sex', y='total_bill', data=tips)
```

```python
import numpy as np
#default estimator is mean, now we used standard deviation (from numpy)
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

#### Count Plot

```python
# Just counts 
sns.countplot(x='sex', data=tips)
```

### Box and Whishker's plot

```python
sns.boxplot(x='day',y='total_bill', data=tips)
# note the whishker (|---- and ----| tells us the range, i.e, min and max value)
# The box's line tells us about the median (i.e half are lower and half are greater than the line)
# The 2 endpoints of the box are again the median of lower and upper half
# In essence the box plot divides the whole data in 4 quartiles 
```

```python
# also it allows to split based on category, eg smokes or not
sns.boxplot(x='day',y='total_bill', data=tips, hue='smoker')
```

#### Violin Plot

```python
sns.violinplot(x='day', y='total_bill', data=tips)
# it kinda shows the frequency (KDE) by the width of the plot
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

#### Strip PLot

```python
sns.stripplot(x='day', y='total_bill', data=tips)
```

```python
# Note by default jitter = True, it scatters the data a bit on x axis to clearly show the points
sns.stripplot(x='day', y='total_bill', data=tips, jitter=False)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex')
```

#### Swarm PLot

```python
# Combine strip and violin
```

```python
# it's violin showing all data
sns.swarmplot(x='day',y='total_bill', data=tips)
```

```python
sns.swarmplot(x='day',y='total_bill', data=tips, hue='sex')
```

```python
# we can stack violin and swarm on top of each other
sns.violinplot(x='day',y='total_bill', data=tips)
sns.swarmplot(x='day',y='total_bill', data=tips, hue='sex', color='cyan')
```

#### Factor Plot

```python
# basically takes another argument as category of plot
sns.catplot(x='day', y='total_bill',data=tips, kind='bar')
```
