```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Mumbai.csv")
```

```python
data.shape
```

```python
data
```

```python
data.sort_values(by='Price', inplace=True)
data.tail()
```

```python
plt.scatter(data['Area'], data['Price'])
plt.siz
plt.xlabel("Area")
plt.ylabel("Price")
```

```python
data = data[["Price","Area"]]
data
```

```python
from scipy import stats
import numpy as np
z = np.abs(stats.zscore(data))
print(z)
```

```python
dataNoOutlier = data[(z < 3).all(axis=1)]
dataNoOutlier
```
